# matrix_mult

> Multiplication between matrices.  
> __Objetive:__ transform the kernel so that each thread takes care of each element in the matrix C (result matrix).

## Usage

### Generate executable

```bash
make
```

### Execute

```bash
./matrix_mult [hA wA/hB wB]
```

For example:
```bash
./matrix_mult 16 16 16
```

## Output

Time A (malloc and memcpy): X s  
Time B (malloc and memcpy): X s  
Time Kernel: X s  
Time C (cudaMemcpyDeviceToHost): X s  
Band width of A: X KB/s  
Band width of B: X KB/s  
Performance Kernel: X GFLOPS/s  
Band width of C: X KB/s  

> [!NOTE]  
> FLOPS = 2 * m * n * k;

## Expectations

Band widths of A and B are faster than C's.
